```python
import requests
import bs4
import os
```

```python
base_url = "https://www.scitechnol.com/archive-electrical-engineering-electronic-technology.php"
result = requests.get(base_url)
result.text
```

```python
all_unique_links =[]
soup = bs4.BeautifulSoup(result.text,"lxml") 
unfiltered_links = (soup.select(".col-xs-12 a "))
for link in unfiltered_links:
    temp_link = link.get('href') 
    if ( temp_link not in all_unique_links):
        all_unique_links.append(temp_link)
        
print(len(all_unique_links))
```

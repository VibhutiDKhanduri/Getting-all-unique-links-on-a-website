```python
def download_pdf(pdf_url,count):
    
    #headers = {
    #"User-Agent": "PostmanRuntime/7.20.1",
    #"Accept": "*/*",
    #"Cache-Control": "no-cache",
    #"Postman-Token": "8eb5df70-4da6-4ba1-a9dd-e68880316cd9,30ac79fa-969b-4a24-8035-26ad1a2650e1",
    #"Host": "https://www.scitechnol.com",
    #"Accept-Encoding": "gzip, deflate",
    #"Connection": "keep-alive",
    #"cache-control": "no-cache",
    #}
    
    params = {
    'switchLocale': 'y',
    'siteEntryPassthrough': 'true'
    }
    
    '''headers = {
    "User-Agent": "PostmanRuntime/7.20.1",
    "Accept": "*/*",
    "Cache-Control": "no-cache",
    "Postman-Token": "8eb5df70-4da6-4ba1-a9dd-e68880316cd9,30ac79fa-969b-4a24-8035-26ad1a2650e1",
    "Host": "https://www.scitechnol.com/peer-review",
    "Accept-Encoding": "gzip, deflate",
    "Connection": "keep-alive",
    "cache-control": "no-cache",
    }
    '''

    modified_url = ""+pdf_url 
    print(modified_url)
    r= requests.get(modified_url,params=params)
    soup= bs4.BeautifulSoup(r.text,'lxml')
    try:
        new_url=soup.select('iframe')[0].get('src')    
        r= requests.get(new_url)
        with open(os.path.join("C:/Users/admin/Thesis Implementation/Engineering Journals/OMICS","File"+str(count)+".pdf"),"wb") as f:
                f.write(r.content)
    except:
        print('Oops there was an error in the file format')
   # with open(os.path.join("C:/Users/admin/Thesis Implementation/Engineering Journals/OMICS","File"+str(count)+".pdf"),"wb") as f:
    #    f.write(r.content)
    
```

```python
#this extracts all unique links of various issues listed in the Journal of Electrical Engineering and Electronic Technology
all_unique_links =[]
soup = bs4.BeautifulSoup(result.text,"lxml") 
unfiltered_links = (soup.select(".nav.row a "))
for link in unfiltered_links:
    temp_link = link.get('href') 
    if ( temp_link not in all_unique_links):
        all_unique_links.append(temp_link)
        print(temp_link,end='\n\n')
        
len(all_unique_links)
```

```python
#gets all the contents from a web page
count= 0
page_count = 0
for base_url in all_unique_links:
    print("\nDownloading Files for page = {}...\n".format(page_count))
    result = requests.get(base_url)
    soup = bs4.BeautifulSoup(result.text,"lxml")
    pdf_url = []
    # type of (soup.find_all('a')[0]) is bs4.element.Tag so we can use it as a dictionary with key value pairs  
    #The find_all method is used for finding out all tags with specified tag name or id
    #for link in
    myList= soup.find_all("a")
    for item in myList: 
        url = item.get('href')
        if url[:26]=='https://www.scitechnol.com'and url[-3:] =='pdf' and url not in pdf_url:
           pdf_url.append(url)                   
    print("Number of pdfs in the page = {}\n".format(len(pdf_url)))
    for file in pdf_url:
        download_pdf(file,count)

        count += 1
    
    page_count +=1
print("Congrats Files Created Successfully !!!")
    
```
